# Temporal Animation of Squid Fishing Hotspots (2000‚Äì2020)
### Spatio-Temporal Shifts, EEZ Enforcement, and High-Seas Fishing in the Southwest Atlantic

## Project Overview
This project explores **spatio-temporal changes in squid fishing activity** in the Southwest Atlantic Ocean through an **animated GIF** generated from representative **QGIS map snapshots**.

The animation uses four key years ‚Äî **2000, 2010, 2011, and 2020** ‚Äî selected to capture the **beginning, transitional phases, and end** of a 20-year squid catch dataset. These snapshots highlight **major spatial regime shifts** in fishing behavior and illustrate how fleets respond over time to **marine governance, enforcement capacity, and ecological dynamics**.

Fishing hotspots are visualized alongside **Argentina‚Äôs Exclusive Economic Zone (EEZ)** boundaries, allowing direct comparison between vessel activity and maritime jurisdiction.

---

## What is an EEZ?
An **Exclusive Economic Zone (EEZ)** is a maritime area extending up to **200 nautical miles from a country‚Äôs coastline**, as defined under the **United Nations Convention on the Law of the Sea (UNCLOS, 1982)**. Within this zone, the coastal state has sovereign rights over natural resources, including fisheries. The red lines in the animated_hotspots.gif and the hotspots maps represent the EEZ boundaries the the countries within that region

- Foreign vessels **cannot legally fish inside an EEZ without authorization**
- Waters beyond the EEZ are considered **high seas**, where international fishing is permitted under international law

**Argentina‚Äôs EEZ was already legally in place well before 2000.**  
Therefore, observed changes in fishing behavior over time are best explained not by the creation of the EEZ, but by **changes in enforcement, surveillance, and compliance**.

---

## Key Observations by Year
### üü¢ 2000 ‚Äì Early Period: Weak Enforcement, High Access
- Foreign fleets (including Taiwan and others not native to the Southwest Atlantic) appear fishing **close to Argentina‚Äôs continental shelf aand had extremely high CPUEs**
- Although the EEZ existed legally, **monitoring and enforcement capacity was limited**
- Fishing likely occurred:
  - Just outside the 200-mile EEZ boundary, or
  - In some cases, illegally inside the EEZ with low risk of interception

This pattern reflects early ‚Äú**open access**‚Äù dynamics near EEZ edges.

---

### üü° 2010 ‚Äì Transitional Period: Moving Offshore
- Fishing activity becames less intense than previous years (lower CPUEs) and shifts **further offshore into deeper waters**
- This likely corresponds with:
  - Improved satellite monitoring (e.g., AIS),
  - Increased naval and coast guard patrols,
  - Growing international attention to Illegal, Unreported, and Unregulated (IUU) fishing

**Interpretation:**  
The EEZ had not changed legally, but **enforcement had become more effective**, encouraging fleets to operate farther from the coast.

---

### üü† 2011 ‚Äì Spatial Shift Southward
- Fishing effort appears to shift **southward toward the Falkland/Malvinas region**
- This change likely reflects a combination of:
  - Ecological factors (squid migration and stock dynamics),
  - Continued pressure to avoid Argentina‚Äôs EEZ,
  - Concentration of activity near politically complex or high-seas-adjacent zones

This year highlights how **fishing fleets adapt spatially** in response to both biology and regulation.

---

### üîµ 2020 ‚Äì Modern Pattern: EEZ Compliance at the Boundary
- Foreign vessels concentrate fishing activity **right along the EEZ boundary**
- Unlike earlier years, they **do not cross into Argentina‚Äôs EEZ**
- This behavior is consistent with:
  - Stronger surveillance and enforcement,
  - Higher penalties and reputational risk,
  - Better vessel tracking and international scrutiny

This pattern reflects the well-known ‚Äú**Mile 201**‚Äù phenomenon, where fleets operate just outside national jurisdiction to legally access migratory species like squid.

---

## Key Takeaways
- The major change in squid catch locations can be due to:
  - Enforcement capacity,
  - Surveillance technology,
  - Compliance behavior of international fleets
- Squid fishing activity increasingly concentrates:
  - Along EEZ boundaries,
  - In high-seas zones adjacent to productive continental shelves

The animated maps visually demonstrate how **policy, enforcement, and ecology intersect spatially** over two decades.

```python
from PIL import Image
import os
```

## Define Input and Output Paths

PNG maps were exported from QGIS and saved under the outputs/maps directory.
The animated GIF will be saved alongside them.

```python
image_dir = "../outputs/maps"
output_gif = os.path.join(image_dir, "animated_hotspots.gif")
```

## Load Map Images

Images are ordered chronologically to reflect temporal progression.

```python
image_files = [
    "hotspots_2000.png",
    "hotspots_2010.png",
    "hotspots_2011.png",
    "hotspots_2020.png"
]

images = [
    Image.open(os.path.join(image_dir, img))
    for img in image_files
]
```

## Create Animated GIF

Frame duration is set to emphasize spatial changes between years.

```python
images[0].save(
    output_gif,
    save_all=True,
    append_images=images[1:],
    duration=2500,   # milliseconds per frame
    loop=0
)

output_gif
```
